```python
import pandas as pd

import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```python
chipotle_data_frame = pd.read_table(url)
chipotle_data_frame
```

```python
filtered = chipotle_data_frame.drop_duplicates(['item_name','quantity'])
one_prod = filtered[filtered.quantity == 1]
price_per_item_df = one_prod[['item_name', 'item_price']]
price_per_item_df
```

1. Sort the values from most to less expensive.

```python
price = lambda price: float(price[1:-1])

price_per_item_df.item_price = price_per_item_df.item_price.apply(price)

price_per_item_df.sort_values(['item_price'], ascending=False)
```

2. What was the qunantity of most expensive item ordered.

```python
price = lambda price : float(price[1:-1])
chipotle_data_frame.item_price = chipotle_data_frame.item_price.apply(price)
chipotle_data_frame.sort_values('item_price', ascending=False).head(1)
```

3. How many times veggie salad bowl ordered

```python
grouped_df = chipotle_data_frame.groupby(by='item_name')
data = grouped_df['item_name'].count()
data['Veggie Salad Bowl']
```
