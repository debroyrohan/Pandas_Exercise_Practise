```python
import pandas as pd
import numpy as np
```

```python
users_data_frame = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
users_data_frame.head()
```

```python
sessions_data_frame = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
sessions_data_frame.head()
```

```python
products_data_frame = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
products_data_frame
```

```python
transactions_data_frame = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
transactions_data_frame
```

1. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

```python
pd.merge(transactions_data_frame, users_data_frame, how='left', left_on='UserID', right_on='UserID')
```

2. Which transactions have UseID not in users?

```python
common_df=pd.merge(transactions_data_frame, users_data_frame, how='inner', left_on='UserID', right_on='UserID')
```

```python
common_df
```

```python
def UserID_not_present_in_common_df(user_id):
    if user_id not in list(common_df.UserID):
        return user_id

index1 = np.where(transactions_data_frame['UserID'] == transactions_data_frame['UserID'].apply(UserID_not_present_in_common_df))
index2 = np.where(transactions_data_frame['UserID'].isnull())
index1 = list(index1[0])
index2 = list(index2[0])
index = list()
index.extend(index1)
index.extend(index2)
transactions_data_frame.loc[index]
```

3. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

```python
pd.merge(transactions_data_frame, users_data_frame, how='inner', left_on='UserID', right_on='UserID')
```

4. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

```python
pd.merge(transactions_data_frame, users_data_frame, how='outer', left_on='UserID', right_on='UserID')
```

5. Determine which sessions occurred on the same day each user registered

```python
def date_not_equal_to_session_date(date):
    if date in sessions_data_frame['SessionID']:
        return date
index = np.where(users_data_frame['Registered'] == users_data_frame['Registered'].apply(date_not_equal_to_session_date))
pd.merge(users_data_frame, sessions_data_frame, how='inner', on= 'UserID').loc[index]
```

6. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
users_data_frame['Key'] = 1
products_data_frame['Key'] = 1
cross_joined_df=pd.merge(users_data_frame, products_data_frame, on='Key').drop('Key', 1)[['UserID', 'ProductID']]
cross_joined_df
```

7. Determine how much quantity of each product was purchased by each user

```python
cross_joined_df['UserID'].astype('float64')
cross_joined_df['ProductID'].astype('float64')
pd.merge(cross_joined_df, transactions_data_frame, how = 'left', on=['UserID', 'ProductID'])[['UserID', 'ProductID', 'Quantity']].fillna(0.0)
```

8. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)

```python
transactions_data_frame['key'] = 1
pd.merge(transactions_data_frame, transactions_data_frame, on='key').drop('key',1)
```

9. Join each user to his/her first occuring transaction in the transactions table

```python
transactions_data_frame.drop('key', axis=1, inplace=True)
users_data_frame.drop('Key', axis=1, inplace=True)  
pd.merge(users_data_frame, transactions_data_frame, how='left', on='UserID').drop_duplicates('UserID')
```
